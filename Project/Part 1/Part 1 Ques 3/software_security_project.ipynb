```python
#loading_csv_file
df = pd.read_csv("/content/projects/data_set/data_set_features_1.csv", on_bad_lines='skip')
#selecting training data samples
training_data = df.sample(frac=0.80)
#dropping training data to create testing data
testing_data = df.drop(training_data.index)
```

```python
df
```

```python
# cleaning data
def get_dataset(data):
  data = data.iloc[:,1:]
  #input vextors of features
  X=data.iloc[:,:-1]
  #output data decision lable for malicous or benign
  y=data.iloc[:,-1]
  return X, y
```

```python
def get_models():
			models = list()
			models.append(LogisticRegression())
			models.append(DecisionTreeClassifier())
			models.append(LinearSVC())
			models.append(SVC())
			models.append(AdaBoostClassifier())
			models.append(RandomForestClassifier())

			return models
```

```python
def evaluate_model(model):
				cv = KFold(n_splits=10, shuffle=True, random_state=1)	
				# get the training input output vectors 
				X_training, y_training = get_dataset(training_data)
				# get the testing input output vectors 
				X_testing, y_testing = get_dataset(testing_data)
				# evaluating the model	
				#predicted_values= cross_val_predict(model, X_testing,y_testing, cv=cv, n_jobs=-1)
				
				#training the model
				model=model.fit(X_training,y_training)
				#predicting_output_for_test_data_set
				predicted_values=model.predict(X_testing)
				#evaluating_f1_scores_for_the_model
				f1 = f1_score(y_testing,predicted_values)
		    #evaluating_recall_scores_for_the_model
				recall=recall_score(y_testing,predicted_values)
		    #evaluating_precision_scores_for_the_model
				precision = precision_score(y_testing,predicted_values)
				#evaluating_confusion_matrix
				confusionmatrix= confusion_matrix(y_testing, predicted_values)
				#print('%s:\nactual_values:\n%s\npredicted_values:\n%s\nconfusion_matrix:\n%s' %(type(model).__name__,df.iloc[:,0][y_testing.index],predicted_values , confusionmatrix))
				# return f1 score
				return  f1,predicted_values,confusionmatrix,df.iloc[:,0][y_testing.index],y_testing,model,precision,recall
```

```python
best_f1=0
for model in get_models():
    predicted_output={}
    # evaluate model using each test condition
    values = evaluate_model(model)
    if isnan(values[0]):
      continue
    if(best_f1<values[0]):
       best_model=values[5]
       best_f1=values[0]
    for i in range(len(values[3].index)):
       if values[1][i] == 0: 
        predicted_output[values[3].array[i]]= "Benign"  
       else: 
        predicted_output[values[3].array[i]]= "Malicious"
       
    with open('data.txt', 'a') as f:
      f.write('%s:\npredicted_values:%s\n\n' %(type(model).__name__,predicted_output))
      f.close()
    print('%s:\n\nTrue_Positive(TP):%s\nFalse_Negative(FN):%s\nFalse_Positive(FP):%s\nTrue_Negative(TN):%s\nf1=%s\nRecall_Score:%s\nPrecision_Score:%s\n' %(type(model).__name__, values[2][0][0],values[2][0][1],values[2][1][0],values[2][1][1],values[0],values[6],values[7]))
    print("********************************************************\n")
```
